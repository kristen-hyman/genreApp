```python
import pandas as pd
```

```python
#import intiial CSV
df = pd.read_csv("rapLyrics.csv", encoding="Utf-8")
df.head()
```

```python
df1 = df[["song", "artist", "genre", "lyrics"]].copy()
df1["lyrics"] = df1['lyrics'].str.lower()
df1.head()
```

```python
import nltk
#nltk.download()
```

```python
df1['tokenized_sents'] = df1.apply(lambda row: nltk.word_tokenize(row['lyrics']), axis=1)
```

```python
df1.head()
```

```python
# Import stopwords with nltk.
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
df1['removedStopwords'] = df1['tokenized_sents'].apply(lambda x: [item for item in x if item not in stop])
df1['bigWords'] = df1['removedStopwords'].apply(lambda x:  [word for word in x if len(word) > 2])
```

```python
df1["bigWords"].head()
```

```python
df1['frequency'] = df1['bigWords'].apply(lambda x: [nltk.FreqDist(x)])
```

```python
df1['frequency'].head(100)
```

```python
df1.head()
```

```python

df1['test'] = df1['bigWords'].apply(lambda x: [nltk.FreqDist(x).most_common(30)])
```
